```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('agaricus-lepiota.data', header = None)
df.columns = ['edibility', 'cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor', 'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color', 'stalk-shape', 'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number', 'ring-type', 'spore-print-color', 'population', 'habitat']
df
```

```python
'''(a) How many features are in this data set?
22

(b) What are the initial names and types of the features?
Categorical Features

(c) Rename the first feature (index 0) to 'edibility' and the sixth feature (index 5) to
'odor'
Done

'''
```

```python
'''
(d) The 6th column lists the various odors of the mushrooms: encoded as 'a', 'c', . . . .
Replace these with the names 'almond', 'creosote', etc. (categories corresponding
to each letter can be found on the website). Also replace the 'edibility' categories
'e' and 'p' with 'edible' and 'poisonous'.
'''

odors = {'a': 'almond', 'l': 'anise', 'c': 'creosote', 'y': 'fishy', 'f': 'foul', 'm': 'musty', 'n': 'none', 'p': 'pungent', 's': 'spicy'}
edibilities = {'p': 'poisonous', 'e': 'edible'}
    
for i,r in df.iterrows():
    odor = r['odor']
    edibility = r['edibility']
    df.at[i, 'odor'] = odors.get(odor)
    df.at[i, 'edibility'] = edibilities.get(edibility)
    
df   
```

```python
'''
(e) Make a contingency table cross-tabulating 'edibility' and 'odor'.
'''
pd.crosstab(df['edibility'], [df['odor']])
```

```python
'''
(f) Which mushroom odors should be avoided, when gathering mushrooms for consumption?

foul, fishy, spicy, pungent, creosote, musty
'''
df[df['edibility'] == 'poisonous'].groupby(['odor'])['odor'].count().sort_values(ascending=False)
```

```python
df[df['edibility'] == 'edible'].groupby(['odor'])['odor'].count().sort_values(ascending=False)
```

```python
'''
(g) What proportion of odorless mushroom samples were safe to eat?
96.6%
'''
df[df['odor'] == 'none'].groupby(['edibility'])['edibility'].count().sort_values()
```
