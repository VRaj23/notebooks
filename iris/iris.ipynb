```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('iris.data', header=None)
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
df.describe()
```

```python
# ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

df['class_color'] = 0
df['petal_area'] = 0
df['sepal_area'] = 0
for i,r in df.iterrows():
    c = r['class']
    v = 0
    if c == 'Iris-setosa':
        v = 'green'
    elif c == 'Iris-versicolor':
        v = 'yellow'
    elif c == 'Iris-virginica':
        v = 'blue'
    df.at[i, 'class_color'] = v
    df.at[i, 'petal_area'] = r['petal_length'] * r['petal_width'] * 3.14
    df.at[i, 'sepal_area'] = r['sepal_length'] * r['sepal_width'] * 3.14
    
df
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
data = {
    'sepal_length': df['sepal_length'],
    'sepal_width': df['sepal_width'],
    'class_color': df['class_color'],
    'petal_area': df['petal_area']
}
ax.scatter('sepal_length', 'sepal_width', c='class_color', s='petal_area', data=data)
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
data = {
    'petal_length': df['petal_length'],
    'petal_width': df['petal_width'],
    'class_color': df['class_color'],
    'sepal_area': df['sepal_area']
}
ax.scatter('petal_length', 'petal_width', c='class_color', s='sepal_area', data=data)
```
