Dataset : [Wine Dataset](https://archive.ics.uci.edu/ml/datasets/wine)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("wine.data", header=None)
df.columns = [
 'class',
 'alcohol',
 'malic_acid',
 'ash',
 'alcalinity_of_ash',
 'magnesium',
 'total_phenols',
 'flavanoids',
 'nonflavanoid_phenols',
 'proanthocyanins',
 'color_intensity',
 'hue',
 'od280/od315',
 'proline']
df
```

```python
df2 = df.copy()
df2['class_color'] = ''
for i,r in df2.iterrows():
    c = r['class'] 
    if c == 1:
        df2.at[i, 'class_color'] = 'green'
    elif c == 2:
        df2.at[i, 'class_color'] = 'yellow'
    elif c == 3:
        df2.at[i, 'class_color'] = 'blue'
        
df2
```

```python
def plot_graph(axs, x_col, y_col):
    data = {
        x_col: df2[x_col],
        y_col: df2[y_col],
        'class_color': df2['class_color']
    }
    axs.set_title(x_col + ' vs ' + y_col)
    axs.set_xlabel(x_col)
    axs.set_ylabel(y_col)
    axs.scatter(x_col, y_col, c='class_color', data=data)
    
cols = ['malic_acid',
 'ash',
 'alcalinity_of_ash',
 'magnesium',
 'total_phenols',
 'flavanoids',
 'nonflavanoid_phenols',
 'proanthocyanins',
 'color_intensity',
 'hue',
 'od280/od315',
 'proline']

fig, axs = plt.subplots(4, 3, figsize=(20, 20))
i=0
j=0
for col in cols:
    plot_graph(axs[i][j], 'alcohol', col)
    j+=1
    if j > 2:
        j = 0
        i+=1
```

```python
X=df.drop(['class'],axis=1)
y=df['class']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X)
xscaled=scaler.transform(X)
X_scaled=pd.DataFrame(xscaled,columns=df.columns[1:])
X_scaled.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score


knn=KNeighborsClassifier(n_neighbors=23)
cv_scores = cross_val_score(knn, X, y, cv=10)
print(cv_scores)
print('cv_scores mean:{}'.format(np.mean(cv_scores)))
```

```python
from sklearn.model_selection import GridSearchCV
#create new a knn model
knn2 = KNeighborsClassifier()
#create a dictionary of all values we want to test for n_neighbors
param_grid = {'n_neighbors': np.arange(3, 25)}
#use gridsearch to test all values for n_neighbors
knn_gscv = GridSearchCV(knn2, param_grid, cv=10)
#fit model to data
knn_gscv.fit(X, y)
knn_gscv.best_params_
```
